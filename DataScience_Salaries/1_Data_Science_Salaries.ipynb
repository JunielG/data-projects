```python
# Histogram
plt.figure(figsize=(18, 8))
sns.histplot(data=ds_salaries, x="salary_in_usd", bins=20, kde=True)
plt.title('histplot - Histogram with KDE')
plt.show()
```

```python
# KDE Plot
plt.figure(figsize=(18, 8))
sns.kdeplot(data=ds_salaries, x="remote_ratio", hue="company_size", fill=True, common_norm=False, alpha=.5)
plt.title('kdeplot - Kernel Density Estimate')
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(18, 8))
sns.scatterplot(data=ds_salaries, x="experience_level", y="salary_in_usd", hue="experience_level")
plt.title('scatterplot - Scatter Plot')
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(18, 8))
sns.scatterplot(data=ds_salaries, x="company_size", y="salary_in_usd", hue="experience_level")
plt.title('scatterplot - Scatter Plot')
plt.show()
```

```python
# Create a custom color palette
colors = ["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"] # "#c2c2f0"]
custom_palette = sns.color_palette(colors)

sns.boxplot(x="company_size",  y="salary_in_usd", hue="experience_level", 
            palette=custom_palette, data=ds_salaries, 
            width=0.6, fliersize=3
)

plt.title("Custom Styled Box Plot", fontsize=12)
plt.xlabel("Company Size", fontsize=12)
plt.ylabel("Total salary_in_usd Amount ($)", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Create a custom color palette
colors = ["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"] # "#c2c2f0"]
custom_palette = sns.color_palette(colors)

sns.boxplot(x="experience_level",  y="salary_in_usd", hue="experience_level", 
            palette=custom_palette, data=ds_salaries, 
            width=0.6, fliersize=3
)

plt.title("Custom Styled Box Plot", fontsize=12)
plt.xlabel("Experience Level", fontsize=12)
plt.ylabel("Total salary_in_usd Amount ($)", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Option 1: Simple approach - Average salary by year and company size
plt.figure(figsize=(12, 6))
salary_by_year_size = ds_salaries.groupby(['work_year', 'company_size'])['salary_in_usd'].mean().reset_index()

# Create line plots for each company size
for size in salary_by_year_size['company_size'].unique():
    data = salary_by_year_size[salary_by_year_size['company_size'] == size]
    plt.plot(data['work_year'], data['salary_in_usd'], marker='o', linewidth=2.5, label=size)

plt.xlabel('Year')
plt.ylabel('Average Salary (USD)')
plt.title('Average Data Science Salaries by Company Size Over Time')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Option 2: Using FacetGrid correctly
plt.figure(figsize=(15, 10))

# Create separate plots for each company size
g = sns.FacetGrid(
    ds_salaries, col="company_size", 
    height=4, aspect=1.2, col_wrap=3
)

# Map box plots or violin plots to show salary distribution by year
g.map_dataframe(sns.boxplot, x="work_year", y="salary_in_usd")

# Customize
g.set_axis_labels("Year", "Salary (USD)")
g.set_titles(col_template="{col_name} Companies")
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle("Data Science Salary Distribution by Company Size", fontsize=16)

# Format x-axis
for ax in g.axes.flat:
    plt.setp(ax.get_xticklabels(), rotation=45)

plt.tight_layout()
plt.show()

# Option 3: Experience level analysis
plt.figure(figsize=(15, 8))

# Create FacetGrid for experience levels
g = sns.FacetGrid(
    ds_salaries, col="experience_level", 
    height=4, aspect=1.2, col_wrap=2
)

# Map line plot showing average salary by year for each experience level
g.map_dataframe(
    lambda **kwargs: sns.lineplot(
        data=kwargs['data'].groupby('work_year')['salary_in_usd'].mean().reset_index(),
        x='work_year', y='salary_in_usd', marker='o', linewidth=2.5
    )
)

# Customize
g.set_axis_labels("Year", "Average Salary (USD)")
g.set_titles(col_template="{col_name} Level")
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle("Average Data Science Salaries by Experience Level", fontsize=16)

# Format x-axis
for ax in g.axes.flat:
    plt.setp(ax.get_xticklabels(), rotation=45)

plt.tight_layout()
plt.show()
```
